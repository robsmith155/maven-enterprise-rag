```python
# Display first page of Item 1A (default chunk size is 10000 characters)
parser.display_section_html('Item 1A', page=2)
```

```python
# Display text content of Item 1 with default line length (100)
parser.display_section_text('Item 1')
```

```python
# Display with custom line length
parser.display_section_text('Item 1', line_length=80)
```

```python
# Display all tables from a section
parser.display_section_tables('Item 7')
```

```python
result = parser.parse_section_with_context('Item 7', table_format='markdown')
```

```python
# Access content in original order
for item in result['content']:
    if item['type'] == 'text':
        print("Text content:", item['content'][:100], "...")  # First 100 chars
    else:  # table
        print("Found table here")

# Access just the tables (same as before)
tables = result['tables']
```

```python
# Display each table
for i, table in enumerate(tables, 1):
    print(f"\nTable {i}:")
    parser.display_table(table)
```
